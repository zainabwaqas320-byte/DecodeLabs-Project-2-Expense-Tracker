```python
# State Initialization (Memory Outside Loop)
total = 0.0
transaction_count = 0
```

```python
# Continuous Audit Loop
while True:
    user_input = input(f"Enter Expense {transaction_count + 1} (or 'quit'): ").strip()

    # 1. The Kill Switch (Sentinel Path)
    if user_input.lower() == 'quit':
        print("\nStopping data entry... Calculating final results.")
        break

    # 2. Digital Poka-Yoke (Type-Safety & Error Handling)
    try:
        expense = float(user_input)

        # Accumulator Pattern (Stateful Ledger)
        total += expense
        transaction_count += 1
        print(f"✔️ Added: Rs. {expense:.2f} | Current Total: Rs. {total:.2f}")

    except ValueError:
        print("❌ Invalid Input! Please enter a valid numerical amount or 'quit'.")
```
